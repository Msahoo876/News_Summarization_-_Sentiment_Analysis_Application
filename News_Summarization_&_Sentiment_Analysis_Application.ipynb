```python
from flask import Flask, request, jsonify, render_template
from transformers import pipeline
import requests
from textblob import TextBlob
```

```python
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")

# Function to fetch news
def get_news(query):
    api_key = ''  # Replace with your actual News API key
    url = (f'https://newsapi.org/v2/everything?q={query}'
           f'&sortBy=publishedAt&language=en&apiKey={api_key}')
    response = requests.get(url)
    return response.json()

# Function to summarize an article
def summarize_article(article_text):
    summary = summarizer(article_text, max_length=50, min_length=25, do_sample=False)
    return summary[0]['summary_text']

# Function to get latest news summary
def get_latest_news_summary(query):
    news_data = get_news(query)
    summaries = []
    for article in news_data['articles'][:5]:  # Get top 5 articles
        article_text = article['content']
        if article_text:
            summary = summarize_article(article_text)
            summaries.append(summary)
    return summaries

# Function to analyze sentiment of a text
def analyze_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'
```

```python

# Example usage
if __name__ == "__main__":
    query = input('Enter Your Message: ')
    summaries = get_latest_news_summary(query)
    if 'weather' in query:
        # response = "The weather is sunny today."
        summaries = get_latest_news_summary('weather')
        response = "\n".join(summaries)
    elif 'news' in query:
        summaries = get_latest_news_summary('news')
        response = "\n".join(summaries)
    else:
        summaries = get_latest_news_summary('query')
        response = "\n".join(summaries)
    print('---------------------------------------------------------------------------------------------------------------------------')
    print(response)
    print('---------------------------------------------------------------------------------------------------------------------------')
```
